```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4]
y= [5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
plt.plot (x,y,marker='o', linestyle = 'dashed',ms = 30, color = 'red', mfc= 'red', mec= 'black', linewidth =3)
plt.show()
plt.legend()
```
